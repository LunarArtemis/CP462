# 6
**Filter data**

```python
left = movies[movies['movie title'].str[-5:-1] > '1990'].iloc[:, 0:3]
left
```

**merge dataframe**

```python
right = ratings.copy()
movie_rating = pd.merge(left,right,left_on="movie id",right_on="movie id")
movie_rating
```

**get result bottom 20 rating movie**

```python
movie_rating = movie_rating.groupby(['movie title'])['rating'].mean().sort_values(ascending=False).tail(20)
movie_rating
# convert to dataframe
movie_rating = pd.DataFrame(movie_rating)
movie_rating
```

# 7
**Save result 6 as CSV**

```python
movie_rating.to_csv('movie_rating.csv', index=False)
```

# 8
**Find a correlation between age and rating**

```python
#Find a correlation between age and rating
corAgeNRating = pd.merge(ratings, users, on='user id', how='inner')
corAgeNRating = corAgeNRating[['age','rating']]
corAgeNRating = corAgeNRating.groupby('age').mean()
corAgeNRating = corAgeNRating.reset_index()
corAgeNRating = corAgeNRating.rename(columns={'rating':'avg_rating'})
corAgeNRating
```
