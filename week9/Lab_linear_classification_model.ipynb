```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, stratify=Y)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

Step 4: สร้าง Linear LogisticRegression model โดยอาศัย features ทั้งหมด

```python
logreg_ovo = LogisticRegression(multi_class="multinomial", max_iter=120)# One vs One
logreg_ovr = LogisticRegression(multi_class="ovr", max_iter=120)# One vs Rest

logreg_ovo.fit(X_train[:, 0:4], Y_train)
```

Step 5: คำนวณค่า log_loss ของ training set และ test set สำหรับโมเดล

```python
predict_prob = logreg_ovo.predict_proba(X_test[:, 0:4])
```

```python
from sklearn.metrics import log_loss
log_loss(Y_test,predict_prob)
```

Step 6: คำนวณค่า mean accuracy  ของ training set และ test set สำหรับโมเดล

```python
logreg_ovo.score(X_test, Y_test)
```

#Question 3: 
* สร้าง Linear LogisticRegression model แบบ OVR และ multimodal โดยอาศัย 2 features (12 โมเดล), 3 features (6 โมเดล) 
* คำนวณค่า log_loss และ ค่า mean accuracy ของ training set และ test set สำหรับแต่ละโมเดล  โดยเปรียบเทียบกับโมเดลที่อาศัย features ทั้งหมด


```python
# Define models for 2 and 3 features
models_2_features = {
    'logreg_ovo2': LogisticRegression(multi_class='multinomial', max_iter=120),
    'logreg_ovr2': LogisticRegression(multi_class='ovr', max_iter=120)
}

models_3_features = {
    'logreg_ovo3': LogisticRegression(multi_class='multinomial', max_iter=120),
    'logreg_ovr3': LogisticRegression(multi_class='ovr', max_iter=120)
}

# Function to calculate log loss_and mean accuracy
def calculate_metrics(model, X_train, X_test, Y_train, Y_test):
    model.fit(X_train, Y_train)
    predict_prob = model.predict_proba(X_test)
    loss = log_loss(Y_test, predict_prob)
    accuracy = model.score(X_test, Y_test)
    return loss, accuracy

# Features definition
features = {
    '2 features': (X_train[:, :2], X_test[:, :2]),
    '3 features': (X_train[:, :3], X_test[:, :3])
}

# Calculate metrics for each model
for feature_name, (X_train_features, X_test_features) in features.items():
    print(f"\nMetrics for {feature_name}:")
    models = models_2_features if '2' in feature_name else models_3_features
    
    for model_name, model in models.items():
        # Calculate metrics
        loss, accuracy = calculate_metrics(model, X_train_features, X_test_features, Y_train, Y_test)
        
        # Print metrics
        print(f'{model_name} - log_loss: {loss}, mean accuracy: {accuracy}')
```
